```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

print('data represent the viscosity of drilling mud taken from 13 manufacturing batches.')

# Load the Excel file into a pandas DataFrame
df = pd.read_excel(r'D:\D.T\geology\G408\point_2_in_3rd_assignment.xlsx')

print(df.columns)

Viscosity_data = df['Viscosity']

# 2. Calculate Mean, Median, Mode for Iron oxide and Aluminum oxide element concentrations
Viscosity_mean = np.mean(Viscosity_data)
Viscosity_median = np.median(Viscosity_data)
Viscosity_mode = df['Viscosity'].mode().values[0]  # Mode might have multiple values, so we select the first one

print("Viscosity_mean =", Viscosity_mean)
print("Viscosity_median =", Viscosity_median)
print("Viscosity_mode =", Viscosity_mode)

# Compute the first quartile and third quartile
quartiles = df.quantile([0.25, 0.75])

# Compute the variance
variance = df.var()

# Compute the standard deviation
std_deviation = df.std()

# Compute the range
data_range = df.max() - df.min()

# Print the results
print("Quartiles:")
print(quartiles)
print("\nVariance:")
print(variance)
print("\nStandard Deviation:")
print(std_deviation)

# Plot a histogram with a bell-shaped curve
sns.histplot(df['Viscosity'], kde=True, color='blue', bins=30)

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram with Bell-Shaped Curve')

# Show the plot
plt.show()

print("yes the data is skewed to the left")

skewness = df['Viscosity'].skew()

print("Coefficient of skewness (B1):", skewness)

print("Since B1 is greater than 1, the right tail is longer, as shown in the bill drawing.")

# Compute the first quartile (25th percentile) and the third quartile (75th percentile)
quartiles = df.quantile([0.25, 0.75])

# Plot the quartiles
quartiles.T.plot(kind='bar', figsize=(7, 4))
plt.title('First Quartile and Third Quartile')
plt.xlabel('')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend(['25th Percentile', '75th Percentile'])
plt.grid(True)
plt.show()
```
